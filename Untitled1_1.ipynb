```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from mpl_toolkits.mplot3d import Axes3D
import pylab
```

```python
def ua(x,t):
    u=x**2/8+np.sin(x-4*t)
    return(u)
```

```python
#n = 51
m = 11
n=(m-1)*3+1
k=(m-1)/(n-1)
h=1/(m-1)
tau=1/(n-1)
#k=tau/h
print(k)
print(h)
print(tau)
```

```python
def makeData (n,m):
    x = np.linspace(0, 1, m)
    t = np.linspace(0, 1, n)
    xgrid, tgrid = np.meshgrid(x, t)

    ugrid = ua(xgrid,tgrid)
    return xgrid, tgrid, ugrid

x, t, u = makeData(n,m)

fig = pylab.figure()
axes = Axes3D(fig)
plt.xlabel('x')
plt.ylabel('t')

axes.plot_surface(x, t, u)

pylab.show()
print(u[n-1])
```

```python
a = [[0] * m for i in range(n)]
tn=np.linspace(0, 1, n)
xn=np.linspace(0, 1, m)
for j in range(len(a[0])):
    a[0][j]=np.sin(xn[j])+((xn[j])**2)/8
for i in range(1,len(a)):
    a[i][0]=-np.sin(4*tn[i])
```

```python
for i in range(len(a)-1):
    for j in range(1,len(a[i])):
        a[i+1][j]=a[i][j]-4*k*(a[i][j]-a[i][j-1])+tau*xn[j]
a=np.array(a)
print(a[n-1])
```

```python
d=a[n-1]-u[n-1]
d.sort
print(d[m-1])
```

```python
dif=[]
mif=[]
m=11
hp=[]
while m<300:
    mif.append(m)
    n=(m-1)*5+1
    k=(m-1)/(n-1)
    h=1/(m-1)
    tau=1/(n-1)
    x, t, u = makeData(n,m)
    a1 = [[0] * m for i in range(n)]
    tn=np.linspace(0, 1, n)
    xn=np.linspace(0, 1, m)
    for j in range(len(a1[0])):
        a1[0][j]=np.sin(xn[j])+((xn[j])**2)/8
    for i in range(1,len(a1)):
        a1[i][0]=-np.sin(4*tn[i])
    for i in range(len(a1)-1):
        for j in range(1,len(a1[i])):
            a1[i+1][j]=a1[i][j]-4*k*(a1[i][j]-a1[i][j-1])+tau*xn[j]
    d=a1[n-1]-u[n-1]
    d.sort
    c=(abs(d[m-1]))
    dif.append(c)
    hp.append(h)
    m=m+10
print(dif)
print(hp)
```

```python
fig = plt.figure(figsize = (12,8))
plt.plot(hp,dif)
plt.grid(True)
plt.title('       .')
plt.xlabel(' ')
plt.ylabel('')
plt.show()
```
