```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from mpl_toolkits.mplot3d import Axes3D
import pylab
I=21.5645*11.605 #потенциал ионизации берем из справочника, элемент He, для неона 24.5874
def calc_func(T, p, g_i = 1, z_a = 1, m = 9.10938356*10e-28, H = 6.582*10e-16):
    #print((2*g_i/z_a) * (T/p) *(m*T/(2*np.pi*H**2))**(3/2) * np.exp(-I/T))
    return((2*g_i/z_a) * (T/p)*(m*T/(2*np.pi*H**2))**(3/2) * np.exp(-I/T))
def alpha(T,P):
    r=calc_func(T,P)
    a=np.sqrt(r/(r+1))
    return(a)
    
```

```python
n=100
def makeData ():
    T = np.linspace(3, 30, n)
    P = np.linspace(0.1, 1, n)
    Tgrid, Pgrid = np.meshgrid(T, P)

    ugrid = alpha(Tgrid,Pgrid)
    return Tgrid, Pgrid, ugrid

T, P, u = makeData()

fig = pylab.figure()
axes = Axes3D(fig)
plt.xlabel('T')
plt.ylabel('P')

axes.plot_surface(T, P, u)

pylab.show()
u=np.array(u)
print(u)
```
